Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?



Ans- The Central Limit Theorem (CLT) is a fundamental concept in statistics that states When random samples of a sufficiently large size are taken from any population with a finite mean and variance, the distribution of the sample means will approximate a normal distribution—regardless of the original population's distribution.
This theorem is critical because it allows statisticians to make inferences about population parameters using the normal distribution, even when the underlying data is not normally distributed. As the sample size increases, the sampling distribution of the sample mean becomes increasingly normal.
The CLT is the foundation for many statistical methods, including confidence intervals and hypothesis testing. It justifies using tools like the z-test and t-test, which assume normality of the sampling distribution. This makes the CLT incredibly powerful and widely applicable across disciplines.
In essence, the CLT allows researchers to work with data from any distribution and still apply parametric methods, provided the sample size is large. This greatly simplifies analysis and increases reliability in decision-making based on sample data.




Question 7: What is the significance of confidence intervals in statistical analysis?



Ans- Confidence intervals (CIs) are crucial in statistical analysis because they provide a range of values that likely contain the true population parameter (such as a mean or proportion), based on sample data. Rather than giving a single estimate, a confidence interval expresses uncertainty in estimation.
For example, a 95% confidence interval for a population mean might be (10.2, 12.8). This means we are 95% confident that the true population mean lies between 10.2 and 12.8.
Significance of Confidence Intervals:
1.	Measure of Precision:
CIs show how precise an estimate is. A narrower interval suggests more precision, often due to larger sample sizes or less variability.
2.	Quantifies Uncertainty:
Rather than assuming the sample estimate is exact, CIs account for sampling variability, giving a realistic range of possible values.
3.	Better Than Point Estimates:
A point estimate (e.g., sample mean) gives no information about variability. CIs provide a more informative and reliable summary.
4.	Used in Hypothesis Testing:
CIs can indicate whether a hypothesized value (e.g., a mean of 0) is plausible. If the hypothesized value falls outside the interval, it may be rejected.




Question 8: What is the concept of expected value in a probability distribution?




Ans- The expected value of a probability distribution is the average or mean value that a random variable is expected to take over many repetitions of an experiment. It represents the long-term average outcome.
For a discrete random variable, the expected value is calculated by multiplying each possible value by its probability and then summing these products. For continuous variables, it involves an integral.
The expected value is important because it provides a single summary measure of the distribution, helping in making predictions and informed decisions. For example, in a fair dice roll, the expected value is 3.5, meaning that over many rolls, the average result will approach 3.5



```python
#Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 50
std_dev = 5
sample_size = 1000

# Generate random numbers
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std_dev = np.std(data)

print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std_dev:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
